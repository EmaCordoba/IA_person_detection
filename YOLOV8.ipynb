```python
from ultralytics import YOLO
import os
# Cargar el modelo preentrenado YOLOv8 para detección
model = YOLO('yolov8n.pt')  # Puedes elegir modelos más grandes como yolov8s, yolov8m

path='data.yaml'
# Entrenar el modelo en el dataset
model.train(data=path, epochs=2, imgsz=640)
```

```python
%matplotlib inline
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Cargar el modelo guardado
model = YOLO("yolov8n.pt")

# Cargar la imagen original
image_path = "./images/test2.jpg"  # Cambia a la ruta de tu imagen
original_image = cv2.imread(image_path)

# Crear una copia para las detecciones
image_with_boxes = original_image.copy()

# Realizar la detección
results = model(image_with_boxes)

# Filtrar las detecciones para mostrar solo personas (ID de clase 'persona' es 0 en COCO)
person_detections = [det for det in results[0].boxes.data if int(det[5]) == 0]

# Dibujar los bounding boxes y mostrar el puntaje de confianza en la copia
for det in person_detections:
    x1, y1, x2, y2, score, class_id = map(int, det[:6])
    cv2.rectangle(image_with_boxes, (x1, y1), (x2, y2), (0, 255, 0), 2)
    cv2.putText(image_with_boxes, f"Confianza: {score:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Convertir BGR a RGB para que las imágenes se muestren correctamente en Matplotlib
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
image_with_boxes_rgb = cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB)

# Mostrar ambas imágenes en una figura con dos subplots
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image_rgb)
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(image_with_boxes_rgb)
plt.title("Detección de Personas")
plt.axis("off")

plt.show()
#cv2.imwrite("imagen_original.jpg", original_image)
#cv2.imwrite("imagen_deteccion.jpg", image_with_boxes)
#print("Las imágenes se han guardado como 'imagen_original.jpg' y 'imagen_deteccion.jpg'.")
```
